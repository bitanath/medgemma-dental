```python
from datasets import load_dataset, Dataset, Features, Image, Value
from pathlib import Path
import json

from huggingface_hub import HfApi
```

```python
%cd dataset_all/
```

```python
records = []
for line in Path("bbox_dataset_simplified.jsonl").read_text().splitlines():
    item = json.loads(line)
    records.append({
        "image": item["image"],                     # provided image
        "prompt": item["prompt"],
        "target": item["target"]
    })

# Define features explicitly
features = Features({
    "image": Image(mode="RGB"),
    "prompt": Value("string"),
    "target": Value("string")
})
```

```python
print(json.dumps(records[0], indent=2))
```

```python
ds = Dataset.from_list(records, features=features)
```

```python
print(ds.features)
```

```python
ds.push_to_hub(
    repo_id="justacoderwhocodes/dental_iopar_panoramic_simplified_bboxes",
    split="train",
    private=True,
    commit_message="Fixed bboxes"
)
```
