```python
from datasets import load_dataset, Dataset, Features, Image, Value
from pathlib import Path
import json

from huggingface_hub import HfApi
```

```python
%cd dataset_classify/
```

```python
records = []
for line in Path("classify_dataset.jsonl").read_text().splitlines():
    item = json.loads(line)
    records.append({
        "image": item["image"],                     # provided image
        "label": item["label"]
    })

# Define features explicitly
features = Features({
    "image": Image(mode="RGB"),
    "label": Value("string")
})
```

```python
print(json.dumps(records[0], indent=2))
```

```python
ds = Dataset.from_list(records, features=features)
```

```python
ds.push_to_hub(
    repo_id="justacoderwhocodes/dental_binary_treatment_classification",
    split="train",
    private=True,
    commit_message="New dataset created"
)
```
